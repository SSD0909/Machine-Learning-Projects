## What is Time series data?

A time series is a sequence of observations taken sequentially in time.

## Examples

* Forecasting the birth rate at all hospitals in a city each year.
* Forecasting product sales in units sold each day for a store.
* Forecasting the number of passengers through a train station each day.
* Forecasting unemployment for a state each quarter.
* Forecasting utilization demand on a server each hour.
* Forecasting the size of the tiger population in a state each breeding season
#### When forecasting time series data, the aim is to estimate how the sequence of observations will continue into the future




## components of time series data

### Stationary
 Mean is constant
 Standard Deviation is constant
 No seasonality/trend
 
### Trend 
The optional and often linear increasing or decreasing behavior of the series over time.

### Seasonality

The optional repeating patterns or cycles of behavior over time.


* A stationary time series is one whose properties do not depend on the time at which the series is observed.


* Thus, time series with trends, or with seasonality, are not stationary  the trend and seasonality will affect the value of the time series at different times. On the other hand, a white noise series is stationary  it does not matter when you observe it, it should look much the same at any point in time.


* In general, a stationary time series will have no predictable patterns in the long-term. Time plots will show the series to be roughly horizontal





## Different Time Series Processes

### White Noise

A series is called white noise if it is random in nature. let {1} denote such a series then it has zero mean [E(1)=0], has a constant variance [V(1)= ^2] and is an uncorrelated [E(t, s) = 0] random variable. in otherwords, Time series that show no autocorrelation are called white noise




### Auto Regressive 

An AR model is one in which yt depends only on its own past values yt-1, yt-2.. etc

### yt =  f(yt-1 , yt-2 ,yt-3...t)

A common representation of AR is as follows 

### Yt = 0 + 1 yt-1 + 2 yt-2 ....+t


### Moving Average

A moving average model is one when yt  depends only on the random error terms which follow a white noise process i.e

### yt = f(t , t-1 , t-2 ,t-3...)

Representation of MA  where it depends on q of its past values is called MA(q)

###  yt =  0 + 1 +  1 t-1 + 2 t-2 +...+ 


the error terms are assumed to be white noise with mean zero and variance is constant

### ARMA

ARMA model considers both AR and MA model reffered as ARMA(p,q)

General Representation of the model

### yt =   0 + 1 yt-1 + 2 yt-2 ....+t + 0 + 1 +  1 t-1 + 2 t-2 +...+ 


###  ARIMA (Auto Regressive Integrated Moving Average)

A series which is not stationary can be made stationary by differencing consecutive values and this is basically called as intergrated.hence ARIMA



Differencing can help stabilise the mean of a time series by removing changes in the level of a time series, and therefore eliminating (or reducing) trend and seasonality.


###  ARIMA and Seasonal ARIMA


## Autoregressive Integrated Moving Averages

The general process for ARIMA models is the following:
* Visualize the Time Series Data
* Make the time series data stationary
* Plot the AutoCorrelation Charts
* Construct the ARIMA Model or Seasonal ARIMA based on the data
* Use the model to make predictions


```python
from pandas import read_csv
from pandas import datetime
from matplotlib import pyplot
from pandas.plotting import autocorrelation_plot
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt
import warnings
warnings.filterwarnings('ignore')



def parser(x):
    return datetime.strptime('190'+x, '%Y-%m')
 
series = read_csv('shampoo.csv', header=0, parse_dates=[0], index_col=0, squeeze=True, date_parser=parser)
print(series.head())
series.plot()
pyplot.show()
autocorrelation_plot(series)
```

```python
series.index = series.index.to_period('M')
# split into train and test sets
X = series.values
size = int(len(X) * 0.66)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
# walk-forward validation
for t in range(len(test)):
    model = ARIMA(history, order=(5,1,0))
    model_fit = model.fit()
    output = model_fit.forecast()
    print(output)
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    print(obs)
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
# evaluate forecasts
rmse = sqrt(mean_squared_error(test, predictions))
print('Test RMSE: %.3f' % rmse)
# plot forecasts against actual outcomes
pyplot.plot(test)
pyplot.plot(predictions, color='red')
pyplot.show()
```
